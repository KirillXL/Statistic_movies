```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

movies_path = '/content/drive/MyDrive/Colab Notebooks/MLdatasets/SHmovies/movies.csv'
```

```python
# Считываем csv файл
movies_df = pd.read_csv(movies_path)
```

```python
# График выпуска фильмов по годам

movies_per_year = movies_df['movieYear'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values)
plt.xlabel('Год')
plt.ylabel('Количество фильмов')
plt.title('Количество фильмов по годам')
plt.show
```
